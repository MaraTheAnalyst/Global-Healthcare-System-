```python
from scipy import stats
```

```python
corr, p_value = stats.pearsonr(df["Healthcare Expenditure (% GDP)"], df["Average Life Expectancy"])
corr, p_value
```

```python
if p_value < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

```python
high_coverage = df[df["Universal Healthcare Coverage (%)"] >= 90]["Infant Mortality Rate (per 1,000)"]
low_coverage = df[df["Universal Healthcare Coverage (%)"] < 90]["Infant Mortality Rate (per 1,000)"]
```

```python
t_stat, p_value = stats.ttest_ind(high_coverage, low_coverage, equal_var=False)
t_stat, p_value
```

```python
single = df[df["Healthcare System Type"] == "Single-payer"]["Patient Satisfaction Score (1-10)"]
multi = df[df["Healthcare System Type"] == "Multi-payer"]["Patient Satisfaction Score (1-10)"]
mixed = df[df["Healthcare System Type"] == "Mixed"]["Patient Satisfaction Score (1-10)"]

stats.f_oneway(single, multi, mixed)
```

```python
from scipy.stats import pearsonr
```

```python
def test_relationship(col_x, col_y):
    valid = df[[col_x, col_y]].dropna()
    corr, p_value = pearsonr(valid[col_x], valid[col_y])
    print(f"\nRelationship between {col_x} and {col_y}")
    print("Correlation:", corr)
    print("P-value:", p_value)
```

```python
test_relationship("Doctors per 1,000 People", "Average Life Expectancy")
```

```python
test_relationship("Doctors per 1,000 People", "Infant Mortality Rate (per 1,000)")
```
